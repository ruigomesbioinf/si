````python
print("Média notas Laboratórios: ", df["Laboratórios"].mean())
print("Média notas Algoritmos Avançados: ", df[" Alg. Avançados"].mean())
print("Média notas Biologia de Sistemas: ", df[" Biologia Sistemas"].mean())
````
